# Introdução
A integração numérica é uma técnica fundamental no cálculo numérico, especialmente quando lidamos com funções cujas primitivas não podem ser expressas em termos de funções elementares, como no caso da função $e^{-x^2}$ presente nesta atividade.Existem diversas técnicas de integração numérica, com destaque para as Fórmulas de Newton-Cotes, particularmente a Regra dos Trapézios e suas variações.

Neste trabalho, abordarei o problema de calcular:

$$ \int_0^1 e^{-x^2} \, dx$$

no intervalo $[0,1]$ com erro absoluto menor que $\frac{1}{6} \times 10^{-4}$, utilizando a Regra dos Trapézios.

# Código

```python
import numpy as np
```

```python
def trapezoidal_rule(f, a, b, n):
    """
    Implementação da Regra dos Trapézios Repetida
    f: função a ser integrada
    a: limite inferior
    b: limite superior
    n: número de subintervalos
    """
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)

    y = f(x)

    integral = (h / 2) * (y[0] + 2 * np.sum(y[1:-1]) + y[-1])
    return integral

    return integral
```

```python
# Função a ser integrada: e^(-x²)
def f(x):
    return np.exp(-x**2)
```

```python
# Parâmetros
a, b = 0, 1
n = 100  # Calculado para erro

# Cálculo da integral
result = trapezoidal_rule(f, a, b, n)
```

```python
print(f"Valor aproximado da integral: {result:.8f}")
```

```python
# Usando quadpack do scipy para um valor mais preciso de referência
from scipy import integrate
valor_real, _ = integrate.quad(f, a, b)
```

```python
print(f"O valor real é: {valor_real:.8f}")
```

```python
erro_absoluto = abs(result - valor_real)
print(f"Erro absoluto: {erro_absoluto:.8f}")
```
